### email draft

```python
def email_drafter(job_company = "Experian", job_title = "Vice President", contact_name = "Jim", contact_title = "HR Director"):
    email_intro = 'Hi ' + contact_name + ' - ' "Hope this message finds you and your family well." + " How are you enjoying your " + contact_title + ' role at ' + job_company + '?' 
    email_pitch = ' I also recently noticed a ' + job_title + ' role at ' + job_company + ' that seems like a good fit for me.'
    email_why_looking = ' As you may have heard, there are big deals happening in the entertainment industry.  Disney offered to buy 20th Century (my current employer) this last December for $52 billion.  As you can imagine such a megadeal brings with it many uncertainties for the acquired company, so I am keeping my options open.'
    email_close = ' Do you know about the role/department? Or would you be able to pass along my resume (attached) to someone in HR? '
    email_why_Im_special = " I've been working on some pretty amazing things here at 20th Century Fox, one data science project was even featured in Variety! "
    email_thank_you = "Thanks " + contact_name + " would love to catch up with you over the phone when you have a moment to learn more about " + job_company + "."
    return email_intro + email_why_looking + email_why_Im_special + email_pitch + email_close + email_thank_you
```
